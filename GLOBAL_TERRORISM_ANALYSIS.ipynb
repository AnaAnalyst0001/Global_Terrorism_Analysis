# **Project Summary**

The Global Terrorism Database (GTD) is an event-level repository with over 200,000 records of terrorist attacks worldwide since 1970. Maintained by the National Consortium for the Study of Terrorism and Responses to Terrorism  (START) at the University of Maryland. They are responsible for periodic updates, emphasizing transparency in database production. GTD emphasizes on several phases of data collection efforts, each relying on publicly available, unclassified source materials. These include media articles and electronic news archives, and to a lesser extent, existing data sets, secondary source materials such as books and journals, and legal documents.
The Global Terrorism Database (GTD) is an open-source database including information on terrorist events around the world from 1970 through 2020 (with additional annual updates planned for the future). Unlike many other event databases, the GTD includes systematic data on domestic as well as transnational and international terrorist incidents that have occurred during this time period and now includes more than 200,000 cases. For each GTD incident, information is available on the date and location of the incident, the weapons used and nature of the target, the number of casualties, and--when identifiable--the group or individual responsible.

To know more, visit here: https://www.start.umd.edu/gtd/

#(E) SUMMARIZING THE ANALYSIS

### 1. Global Terrorism Overview: Incidents, contributions, and Casualties by Country

```python


# Group by 'country_txt' and calculate the aggregations
country_summary = data.groupby('country_txt').agg(
    incidents=('country_txt', 'count'),
    percentage=('country_txt', lambda x: len(x) / len(data) * 100),
    total_killed=('nkill', 'sum'),
    total_injured=('nwound', 'sum')
).reset_index()

# Rename columns for clarity
country_summary.columns = ['Country', 'Incidents', 'Percentage', 'Total Killed', 'Total Injured']

# Sort the table by number of instances in descending order
country_summary = country_summary.sort_values(by='Incidents', ascending=False)

# Reset the index for a cleaner output
country_summary.reset_index(drop=True, inplace=True)

# Print the summary table
country_summary.head(10)
```

**1. Visualizing number of attacks per country**

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'country_summary'

# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world shapefile with your country_summary DataFrame
world = world.merge(country_summary, how='left', left_on='name', right_on='Country')

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=0.5, color='black')  # Plot country boundaries
world.plot(column='Incidents', ax=ax, legend=True, cmap='YlOrRd', edgecolor='black', linewidth=0.5)
ax.set_title('Number of Incidents by Country')
plt.show()
```

A Choropleth map is used to visually represent spatial variations in data through shading or coloring geographic areas. In this visualization, we have depicted **Attacks per Country.** A the number of attacks increases, shading becomes darker.

We can indicate that, **Iraq, Pakistan, Afghanistan, India and Colombia** witnessed the highest number of attacks, in 47 years (1970-2017).

**11. What are the top perpatrator gangs in each region?**

```python
grouped_data = data.groupby(['region_txt', 'gname']).size().reset_index(name='count')

top_gname_by_region = grouped_data.groupby('region_txt').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

top_gname_by_region
```

```python
# Create a horizontal bar chart using Plotly Express
fig = px.bar(top_gname_by_region, x='count', y='gname', color='region_txt',
             title='Top 5 Terrorist Groups by Region',
             labels={'count': 'Count'}, height=1000)
fig.show()
```

An **unknown gang** caused highest attacks in **middle east & North african** region, folllowed by **south Asia**.
